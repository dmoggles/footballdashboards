```python
from footballdashboards.dashboard.shotplotdashboard import ShotPlotDashboard
from examples.sample_datasource import SampleDatasource
import pandas as pd
%matplotlib inline
```

```python
dash = ShotPlotDashboard(SampleDatasource())
table_data = pd.DataFrame({'Player':['Player 1', 'Player 2', 'Player 3','Player 4'], 'xA':[0.1, 0.2, 0.3, 0.4], 'xG':[0.1, 0.2, 0.3, 0.4]})
```

```python
fig, axes = dash.plot_player_seasons_leagues("Lionel Messi", [2022], ["La Liga","Champions League"])
inset_ax = axes["pitch"].inset_axes([0.04, 0.05, 0.2, 0.2])
inset_ax.axis("off")
table = inset_ax.table(
    cellText=table_data.values, cellLoc='left', colLabels=table_data.columns, loc='center', cellColours=[['#000011'] * len(table_data.columns)] * len(table_data), colColours=['#000055'] * len(table_data.columns)
)
table.properties()

table_props = table.properties()
table_cells = table_props['children']
for cell in table_cells: 
    
    cell.get_text().set_color('ivory')
```
