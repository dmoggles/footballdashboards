```python
from examples.sample_datasource import SampleDatasource
from footballdashboards.dashboard.scatterdashboard import ScatterDashboard, ScatterTexts, ScatterWithQuadrantsDashboard, ScatterWithRegressionDashboard, Dashboard
```

```python
dashboard = ScatterWithQuadrantsDashboard(
    SampleDatasource(), 
    x_column="Turnovers Per 100 Touches",
    y_column="Shots Created Per 100 Touches", 
    texts=ScatterTexts(title='Creating Shots, But At What Cost?', subtitle='Shot creation and possession loss on a per touch basis')
)
dashboard.bottom_boundary_multiplier = 1.2
dashboard.name_annotation_threshold = 0.1
dashboard.quadrant_texts = {
    'topright': 'High Shot Creation, Low Possession Loss',
    'topleft': 'Low Shot Creation, Low Possession Loss',
    'bottomright': 'High Shot Creation, High Possession Loss',
    'bottomleft': 'Low Shot Creation, High Possession Loss'
}
dashboard.top_boundary_multiplier = 1.2
```

```python
dashboard.plot_seasons_leagues(seasons=2022, leagues=['Premier League','Serie A','Bundesliga','La Liga','Ligue 1','Primeira Liga','Eredivisie'])
```

```python
dashboard.describe_adjustable_fields()
```
